```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
k = 50
x = np.array([100, 1000, 5000, 10000, 15000, 20000, 25000, 30000])
y = np.array([0.02586, 0.2310, 1.258, 2.835, 4.444, 5.955, 7.339, 8.979])
y = y * 1000

ks = k * x
avg_constant = sum(y / ks) / len(y)

```

```python
plt.plot(x, y, marker='o', label='Empirical Time')
plt.plot(x, ks * avg_constant, label='Theoretical Time (O(s) at constant K | KS)')
plt.xlabel('Sequence Length N | Smallest Size S')
plt.ylabel('Time (ms)')
plt.title('Empirical vs Theoretical Time Complexity of Banded Alignment')
plt.legend()
plt.show()

print(f'Average constant factor: {avg_constant}')
```
