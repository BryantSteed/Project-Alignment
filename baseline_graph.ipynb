```python
import matplotlib.pyplot as plt
import numpy as np
from alignment import align
```

```python
x = np.array([500, 1000, 1500, 2000, 2500, 3000])
y = np.array([0.671, 2.618, 6.480, 11.02, 17.72, 26.05])
y = y * 1000
```

```python
squared = x ** 2
squared_constant = sum(y / squared) / len(y)
```

```python
plt.plot(x, y, marker='o', label='Empirical Time')
plt.plot(x, squared * squared_constant, label='Theoretical Time (O(N^2 | MN))')
plt.xlabel('Input Size (N)')
plt.ylabel('Time (ms)')
plt.title('Unrestricted Empirical vs Theoretical Time Complexity')
plt.legend()
plt.show()

print("Constant of Proportionality:", squared_constant)
```
